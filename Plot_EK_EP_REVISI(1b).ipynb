```python
#======================Energy==============================
def potential_energy(m,h):
    """
    Calculates potential energy of a body of mass m at height h
    Args:
    -m: Mass of the object
    -h: Height of the object
    Returns:
    -p_e: Potential energy
    """
    g = 9.81 #m/s^2
    p_e = m*g*h
    return p_e

def kinetic_energy(m,h):
    """
    Calculates kinetic energy of a body
    Args:
    -m: Mass of the object
    -v: Velocity of the object
    Returns:
    -k_e: Kinetic Energy
    """
    g = 9.81 #m/s^2
    v=(2*g*h)**0.5
    k_e = (1/2)*m*(v**2)
    return k_e


def velocity(h):
    """
    Calculates velocity of a freely falling object
    Args:
    -t: time
    Returns:
    -vel: free fall velocity
    """
    g = 9.81 #m/s^2
    vel = (2*g*h)**0.5
    return vel

def position(h):
    """
    Calculates position of a freely falling object
    Args:
    -vel: Velocity of the object at a given time
    -t: time
    Returns:
    -y: Position of the body
    """
    y = 50
    t = (2*h/g)**0.5
    v = (2*g*h)**0.5
    y = y - (v*t)
    return y
```
